```
Pole Zero Plot of Low Pass Filter
```



````python
import matplotlib.pyplot as plt
from scipy import signal

# Transfer function coefficients
c1 = 2.5e-6
c2 = 2.5e-6
R1 = 1000
R2 = 1000

# Define the transfer function coefficients
numerator = [1]
denominator = [c1*c2*R1*R2, c1*(R1+R2), 1]

# Create a system
system = signal.TransferFunction(numerator, denominator)

# Get the poles and zeros
poles = system.poles
zeros = system.zeros

# Plot the pole-zero plot
plt.scatter(poles.real, poles.imag, marker='x', color='red', label='Poles')
plt.scatter(zeros.real, zeros.imag, marker='o', color='blue', label='Zeros')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Pole-Zero Plot')
plt.legend()
plt.grid(True)
plt.show()
````

Pole zero plot for band Pass Filter

````python
import matplotlib.pyplot as plt
from scipy import signal

# Transfer function coefficients
c1 = 47e-3
c2 = 2.5e-6
R1 = 2000
R2 = 1000

# Define the transfer function coefficients
numerator = [-c1]
denominator = [c2, R1*c1*c2, R1*c1 + R2*c1, R2]

# Create a system
system = signal.TransferFunction(numerator, denominator)

# Get the poles and zeros
poles = system.poles
zeros = system.zeros

# Plot the pole-zero plot
plt.scatter(poles.real, poles.imag, marker='x', color='red', label='Poles')
plt.scatter(zeros.real, zeros.imag, marker='o', color='blue', label='Zeros')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.title('Pole-Zero Plot')
plt.legend()
plt.grid(True)
plt.show()
````
