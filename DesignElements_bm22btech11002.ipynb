Input Output Characteristics of Mosfet

```python
import numpy as np
import matplotlib.pyplot as plt

K = 0.02
Vth = 1.0
Vdd = 10.0
Rd = 12.5

Vi_values = np.linspace(0, 2, 100)

def calculate_Id(Vi):
    return K * (Vi - Vth)**2

def calculate_Vo(Vi):
    if Vi < Vth:
        return Vdd
    else:
        Id = calculate_Id(Vi)
        return Vdd - Id * Rd

Vo_values = [calculate_Vo(Vi) for Vi in Vi_values]

plt.figure(figsize=(8, 6))
plt.plot(Vi_values, Vo_values, label="Vo vs. Vi")
plt.xlabel("Input Voltage (Vi) [V]")
plt.ylabel("Output Voltage (Vo) [V]")
plt.title("MOSFET Vo vs. Vi Characteristic")
plt.legend()
plt.grid(True)
plt.show()
```

Output Characteristics for Sinosuidal Input

```python
import numpy as np
import matplotlib.pyplot as plt

K = 0.02
Vth = 1.0
Vdd = 10.0
Rd = 12.5

t = np.linspace(0, 4 * np.pi, 1000)
Vi_values = 5 * np.sin(t) + 6

def calculate_Id(Vi):
    return (K * (Vi - Vth)**2)/2

def calculate_Vo(Vi):
    if Vi < Vth:
        return Vdd
    else:
        Id = calculate_Id(Vi)
        return Vdd - Id * Rd

Vo_values = [calculate_Vo(Vi) for Vi in Vi_values]

plt.figure(figsize=(10, 6))
plt.plot(t, Vi_values, label="Vi (Input)")
plt.plot(t, Vo_values, label="Vo (Output)")
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.title("MOSFET Vo vs. Vi with Sinusoidal Input")
plt.legend()
plt.grid(True)
plt.show()
```

Load Line Characteristics

```python
import numpy as np
import matplotlib.pyplot as plt

K = 0.02
Vth = 1.0
Vdd = 10.0
Rl = 12.5
Vds_max = Vdd - Vth

Vds_values = np.linspace(0, Vds_max, 100)
Id_values = [(Vdd - Vds) / Rl if Vds <= Vds_max else 0 for Vds in Vds_values]

plt.figure(figsize=(8, 6))
plt.plot(Vds_values, Id_values, label="Load Line")
plt.xlabel("Drain-Source Voltage (Vds) [V]")
plt.ylabel("Drain Current (Id) [A]")
plt.title("MOSFET Load Line")
plt.legend()
plt.grid(True)
plt.show()
```
